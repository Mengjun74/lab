```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
#since the min sum value for two dice is 2 max value is 12 
p = []
for i in range(2,13):
    if(i <= 7):
        curr_p = (i-1)/6 * 1/6
        p.append(curr_p)
    if(i > 7):
        curr_p = (13-i)/6 * 1/6
        p.append(curr_p)
p
```

```python
x = np.linspace(2,12,11)
plt.plot(x,p)
plt.title('rough plot for convoluition operator')
plt.ylabel('Porbability')
plt.xlabel('X')
```

Calculate the mean and the variance of the distribution in problem 2. Hint: this is surprisingly tricky, make sure your result makes sense.

```python
p_mean = np.mean(p)
p_std = np.std(p)
print(p_mean,p_std)
```

```python
#max sum is 60, min sum is 10, hence, it should run 50 times
p_10 = []
for i in range(1,52):
    if(i <= 6):
        p_10.append(i/6 * (1/6)**8 * 1/6)
    elif(i <= 12):
        p_10.append((i-6)/6 * (1/6)**7* 1/6)
    elif(i <= 18):
        p_10.append((i-12)/6 * (1/6)**6* 1/6)
    elif(i <= 24):
        p_10.append((i-18)/6 * (1/6)**5* 1/6)
    elif(i <= 30):
        p_10.append((i-24)/6 * (1/6)**4* 1/6)
    elif(i <= 36):
        p_10.append((i-30)/6 * (1/6)**5* 1/6)
    elif(i <= 42):
        p_10.append((i-36)/6 * (1/6)**6* 1/6)
    elif(i <= 48):
        p_10.append((i-42)/6 * (1/6)**7* 1/6)
    elif(i <= 52):
        p_10.append((i-48)/6 * (1/6)**8* 1/6)
p_10
```

```python
x = np.linspace(10,60,60)
plt.plot(x,p_10)
plt.title('rough plot for convoluition operator')
plt.ylabel('Porbability')
plt.xlabel('X')
```
